```python
##Checking for missing values
for col in df2.columns:
    missing_values = df2[col].isnull().sum()
    if missing_values > 0:
        print("!!!The {} column had {} missing values".format(col, missing_values))
    else:
        print("The {} column has no missing values".format(col, missing_values))
```
